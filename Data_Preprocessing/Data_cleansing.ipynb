```python
#    
from pandas import read_csv

try:
    from rich.console import Console
    from rich.syntax import Syntax
except ImportError:
    print("'rich'   .")
    print('          .')
    print('       .')

    def print_code(code):
        print()
        for line in code.split('\n'):
            print('    ' + line)
        print()
else:
    print("'rich'   .")
    print('    .')
    console = Console()

    def print_code(code):
        print()
        for line in code.split('\n'):
            print('    ', end='')
            console.print(Syntax(line, 'python', theme='ansi_light'))
        print()


classes = ['',
           ' ',
           ' ',
           '  ',
           '-1 ( )',
           '-2 ( )',
           '-3 ( )',
           ' ',
           '']
                                
index = -1
print("\n  'Enter'    ,    ")

while True:
    try:
        index = int(input('\n\n : '))
    except ValueError:
        index += 1

    try:
        data = train.loc[index]
    except KeyError:
        break
    else:
        class_idx, problem, code, answer = data

    print(f'{index} , {class_idx} , {classes[class_idx]}')
    print(problem)
    print_code(code)
    print(': ', end='')
    exec(code)
    print(':', answer)
```
